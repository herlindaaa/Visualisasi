```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("mxmh_survey_results.csv")
data.head()
```

```python
data.info()
```

```python
data['Timestamp'].unique()
```

```python
latest_date = '9/11/2022 01:55:20'
data[data['Timestamp']==latest_date]
```

```python
unique_streaming = data[data.Anxiety>0]['Primary streaming service'].unique()
unique_streaming
```

```python
primary_anxiety_case = []
for i in unique_streaming:
  primary_anxiety_case.append(data[data.Anxiety>0][data['Primary streaming service']==i])
```

```python
unique_genre = data['Fav genre'][data.Anxiety>1].unique()
unique_genre
```

```python
genre_anxiety_case = []
for i in unique_genre:
  genre_anxiety_case.append(data[data.Anxiety>0][data['Fav genre']==i])
```

```python
nan_indices = []
for i in range(len(unique_genre)):
  if type(unique_genre[i])==float:
    nan_indices.append(i)
```

```python
unique_genre = list(unique_genre)
genre_anxiety_case = list(genre_anxiety_case)
```

```python
for i in nan_indices:
  unique_genre.pop(i)
  genre_anxiety_case.pop(i)
unique_genre
```
